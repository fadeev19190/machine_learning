```python
data_all.dropna(inplace=True)
```

```python
data_all_parameters_list
```

```python
X = data_all[data_all_parameters_list]
y = data_all['yield']
```

```python
data_all.info()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import f1_score

yield_model = RandomForestRegressor(random_state=1)
yield_model.fit(X, y)
```

```python
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

yield_model.fit(train_X, train_y)
pred = yield_model.predict(val_X)

r2_score = r2_score(val_y, pred)
print('R^2 score of Random Forest Regressor is:', (r2_score * 100), '%')

# mse_rfr = mean_squared_error(val_y, pred)
```

```python
X_test = df_test[data_all_parameters_list]

pred_test = yield_model.predict(X_test)
```

```python
output = pd.DataFrame({'id': df_test.id,
                       'yield': pred_test})
output.to_csv('submission_yield.csv', index=False)
```
