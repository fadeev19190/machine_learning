```python
import numpy as np
import pandas as pd

train_ds = pd.read_csv('train.csv')
train_ds['Age'] = train_ds['Age'].fillna(30)
# train_ds.dropna(subset="Age", inplace=True)
train_ds
```

```python
train_ds.loc[:, 'Sex'] = train_ds['Sex'].replace({'male': 0, 'female': 1})
```

```python
features = ['Pclass', 'Fare', 'Sex', 'SibSp', 'Parch', 'Age']

y = train_ds.Survived

X = train_ds[features]

from sklearn.ensemble import RandomForestRegressor

titanic_model = RandomForestRegressor(random_state=1)
titanic_model.fit(X, y)
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

titanic_model.fit(train_X, train_y)

titanic_pred = titanic_model.predict(val_X)
binary_predictions_with_age = [1 if p >= 0.5 else 0 for p in titanic_pred]
```

```python
from sklearn.metrics import accuracy_score

titanic_accuracy_rfr = accuracy_score(binary_predictions_with_age, val_y)
titanic_accuracy_rfr
```

```python
test_ds = pd.read_csv('test.csv')
test_ds.loc[:, 'Sex'] = test_ds['Sex'].replace({'male': 0, 'female': 1})
test_ds['Age'] = test_ds['Age'].fillna(30)
test_ds['Fare'] = test_ds['Fare'].fillna(8)

test_ds
# test_ds = test_ds.dropna()
```

```python
mean_age_by_sex = test_ds.groupby('Sex')['Age'].mean()
mean_age_by_sex
```

```python
features_without_age = ['Pclass', 'Fare', 'Sex', 'SibSp', 'Parch', 'Age']
X_test = test_ds[features_without_age]

titanic_pred_test = titanic_model.predict(X_test)
binary_predictions = [1 if p >= 0.5 else 0 for p in titanic_pred_test]
```

```python
output = pd.DataFrame({'PassengerId': test_ds.PassengerId,
                       'Survived': binary_predictions})
output.to_csv('submission1.csv', index=False)

sub = pd.read_csv('submission1.csv')
sub
```
