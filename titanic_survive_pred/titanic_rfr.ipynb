```python
train_ds_without_age = train_ds.dropna(axis=1)
train_ds_without_age
```

```python
train_ds_without_age.loc[:, 'Sex'] = train_ds_without_age['Sex'].replace({'male': 0, 'female': 1})

train_ds_without_age
```

```python
features_without_age = ['Pclass', 'Fare', 'Sex', 'SibSp', 'Parch']
y = train_ds_without_age.Survived

X = train_ds_without_age[features_without_age]

X.info()
```

```python
women = train_ds.loc[train_ds.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% of women who survived:", rate_women)
```

```python
men = train_ds.loc[train_ds.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% of men who survived:", rate_men)
```

```python
from sklearn.ensemble import RandomForestRegressor

titanic_model_without_age = RandomForestRegressor(random_state=1)
titanic_model_without_age.fit(X, y)
```

```python
from sklearn.model_selection import train_test_split

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

titanic_model_without_age.fit(train_X, train_y)

titanic_pred = titanic_model_without_age.predict(val_X)
binary_predictions = [1 if p >= 0.5 else 0 for p in titanic_pred]
```

```python
from sklearn.metrics import accuracy_score

titanic_accuracy_without_age_rf = accuracy_score(binary_predictions, val_y)
titanic_accuracy_without_age_rf
```

```python
test_ds = pd.read_csv('test.csv')
test_ds.loc[:, 'Sex'] = test_ds['Sex'].replace({'male': 0, 'female': 1})
test_ds['Fare'] = test_ds['Fare'].fillna(8)
test_ds = test_ds.drop('Age', axis=1)
test_ds = test_ds.drop('Cabin', axis=1)
test_ds = test_ds.dropna()
```
