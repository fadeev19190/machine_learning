```python
categorical_cols = house_data_train.select_dtypes(include=['object']).columns

# apply one-hot encoding to each categorical variable
for col in categorical_cols:
    one_hot = pd.get_dummies(house_data_train[col], prefix=col)
    data = pd.concat([house_data_train, one_hot], axis=1)
    house_data_train.drop(col, axis=1, inplace=True)
```

```python
house_data_train.head(10)
```

```python
features = [col for col in house_data_train.columns if col not in ['Id', 'SaleType', 'SaleCondition', 'SalePrice']]

X = house_data_train[features]
y = house_data_train['SalePrice']
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split

house_price_model_rfr = RandomForestRegressor(random_state=1)

train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)
house_price_model_rfr.fit(train_X, train_y)

pred_house_price_model_rfr = house_price_model_rfr.predict(val_X)

r2_pred_house_price_model_rfr = r2_score(val_y, pred_house_price_model_rfr)

r2_pred_house_price_model_rfr
```
